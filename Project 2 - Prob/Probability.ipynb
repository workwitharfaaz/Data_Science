```python
# TOSS PROBABILITY

import random

tosses = 10000

results = [random.choice([0, 1]) for _ in range(tosses)]

count_0 = results.count(0)
count_1 = results.count(1)


total = len(results)
percent_0 = (count_0 / total) * 100
percent_1 = (count_1 / total) * 100

print(f"Heads: {percent_0:.2f}%")
print(f"Tails: {percent_1:.2f}%")



#we import random then we create a variable "tosses"
#then we create a list results that has values 0s and 1s 
#"tosses" times. The probability of the value being 0 or 
#1 is 1/2 because we used random.choice() and gave
#it two choices [0, 1].
#Then we count 0s and 1s and we divide them both with 
#the length of results to get their probabilities.
```

```python
# PROBABILITY OF SUM EQUALS 7, 2 DICE

import random

throws = 10
sum7 = 0

for _ in range(throws):
    die1 = random.choice([1, 2, 3, 4, 5, 6])
    die2 = random.choice([1, 2, 3, 4, 5, 6])
    if die1 + die2 == 7:
        sum7 += 1

percent_sum7 = (sum7 / throws) * 100

print(f"Probability of sum = 7 with two dice: {percent_sum7:.2f}%")



#we import random then we create variables "throws"
#and "sum7" then we create a for loop that simulates 2
#dice each having 6 values. Then we create an if 
#statement that checks if the sum of the value from 
#die1 and die2 = 7. If it is then we add 1 to sum7 
#variable. Then we divide sum7 by the number of "throws"
#to get the probability.
```

```python
# PROBABILITY OF 6, FAIR DIE

import random

throws = 10

results = [random.choice([1, 2, 3, 4, 5, 6]) for _ in range(throws)]

count_6 = results.count(6)


total = len(results)
percent_6 = (count_6 / total) * 100

print(f"Probability of 6: {percent_6:.2f}%")



#we import random then we create variables "throws"
#and a list "results" that has values 1 to 6 "throws" times.
#Then we create a variable that has the count of how many 
#times we got 6 in the results list. we divide the count_6
#variable by the total number of results to get the 
#probability of 6.
```

```python
# PROBABILITY OF 6, FAIR DIE

import random

throws = 1000
count_6 = 0

for _ in range(throws):
    result = random.choice([1, 2, 3, 4, 5, 6])
    if result == 6:
        count_6 += 1
    else:
     continue


percent_6 = (count_6 / throws) * 100

print(f"Probability of 6: {percent_6:.2f}%")



#This is same as previous but without the results list.
#we import random then we create variables "throws"
#and "count_6" and for loop that has "result" variable
#that has a single value from 1 to 6.
#If result equals 6 then we increment "count_6" by 1,
#if not then we continue. After this we divide the count_6
#variable by the throws to get the probability of 6.
```

```python
# Conditional Probability and Bayes' Theorem

import random

draw = 1000

drawn_balls = []


for i in range(draw):
    ball = random.choice(range(1,21))
    if ball <6:
        drawn_balls.append("Red")
    elif ball <13:
        drawn_balls.append("Blue")
    else:
        drawn_balls.append("Green")

count_previous_blue = 0
count_red_after_blue = 0

for i in range(1, draw):  #we start from second draw
    if drawn_balls[i-1] == "Blue":
        count_previous_blue += 1
        if drawn_balls[i] == "Red":
            count_red_after_blue += 1


prob_red_given_blue = (count_red_after_blue / count_previous_blue) * 100
print(f"The probability of red ball if the previous ball was blue = {prob_red_given_blue:.2f}%")


#we import random. we create a variable "draw" and list "drawn balls"
#we create a for loop, a "ball variable" that can have a number, 1 to 20. 
#If the number is less than 6 then we append "Red" to drawn_balls,
#else if less than 13, we append "Blue" to drawn_balls,
#else we append "Green. To compute the conditional probability
#we create a variable count_previous_blue
#and count_red_after_blue. For loop that starts from 2nd draw
#if drawn_balls previous value is Blue then we increment 1 to count_previous_blue
#if current value is Red we increment 1 to count_red_after_blue.
#Then we divide count_red_after_blue by count_previous_blue to get probability
```

```python
# Random Variables and Discrete Probability

import random
import numpy as np


values = [1,2,3]
probs= [0.25, 0.35, 0.40]

sample = random.choices(values, weights=probs, k=1000)

mean = np.mean(sample)
var = np.var(sample)
sd = np.std(sample)


print("Mean:",mean)
print("variance:",var)
print("Standard Deviation:",sd)


# we import random and numpy
# then we create lists "values" and "probs"
# and "sample" is 1000 values of 1s,2s and 3s with
# 0.25, 0.35 and 0.40 probabilities respectively.
# Now we calculate mean, variance and standard deviation
# of the sample with the help of numpy 
# library .mean, .var and .std.

```

```python
import numpy as np
import matplotlib.pyplot as plt

samples = np.random.exponential(scale=5, size=2000)

plt.hist(samples, bins=30, density=True, alpha=0.6, color="skyblue", edgecolor="black", label="Histogram (samples)")

x = np.linspace(0, np.max(samples),200)
lambda_param = 1 / 5
pdf = lambda_param * np.exp(-lambda_param * x)

plt.plot(x, pdf, 'r-', linewidth=4, label="PDF")

plt.title("Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()


# we import numpy and matplotlib.
# then we create a samples variable that
# has 2000 random samples from
# exponential distribution with a mean of 5
# then we plot a histogram with 30 bins.
# then we generate 200 equally spaced values
# from 0 to largest sample
# we compute the exponential PDF formula with 
# lambda = 1/mean or 1/5
# then we plot it as red
# then we plot labels and legend
```
