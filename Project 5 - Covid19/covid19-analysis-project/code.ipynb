```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid19_Confirmed_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(["Lat", "Long"], axis =1, inplace = True)
```

```python
df.drop(["Province/State"], axis =1, inplace = True)
```

```python
df.head()
```

```python
df_aggregated = df.groupby("Country/Region").sum()
```

```python
df_aggregated.head()
```

```python
df_aggregated.shape
```

```python
df_aggregated.loc['India'].plot()
df_aggregated.loc['China'].plot()
plt.legend()
```

```python
df_aggregated.loc["China"].plot()
```

```python
df_aggregated.loc["China"][:3].plot()
```

```python
# first derivative
df_aggregated.loc["China"].diff().plot()
df_aggregated.loc["India"].diff().plot()
df_aggregated.loc["US"].diff().plot()
plt.legend()
```

```python
# maximum infection rate for all the countries
countries = list(df_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(df_aggregated.loc[c].diff().max())
df_aggregated["max_infection_rate"] = max_infection_rates
```

```python
df_aggregated.head()
```

```python
# new dataset with only useful columns
corona_data = pd.DataFrame(df_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

```python
# Happiness dataset
happiness_data = pd.read_csv("worldwide_happiness_report.csv")
happiness_data.head()
```

```python
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
happiness_data.drop(useless_cols, axis=1, inplace=True)
happiness_data.head()
```

```python
happiness_data.set_index("Country or region", inplace=True)
happiness_data.head()
```

```python
# Join two datasets - Covid19 and happiness Dataset
```

```python
# Covid19 dataset
corona_data.head()
```

```python
corona_data.shape
```

```python
# World Happiness Dataset
happiness_data.head()
```

```python
happiness_data.shape
```

```python
# Inner JOIN to join the dataset
data = corona_data.join(happiness_data, how="inner")
data.head()
```

```python
data.shape
```

```python
data.corr()
```

```python
data.head()
```

```python
# GDP vs maximum infection rate
x1 = data["GDP per capita"]
y1 = data["max_infection_rate"]
sns.regplot(x = x1,y = np.log(y1))
```

```python
# Plotting Social Support vs maximum infection rate
x1 = data["Social support"]
y1 = data["max_infection_rate"]
sns.regplot(x = x1,y = np.log(y1))
```

```python
# Plotting Health life expectancy vs maximum infection rate
x1 = data["Healthy life expectancy"]
y1 = data["max_infection_rate"]
sns.regplot(x = x1,y = np.log(y1))
```
