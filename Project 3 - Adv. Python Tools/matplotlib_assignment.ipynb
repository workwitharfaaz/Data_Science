```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

we create two lists x and y

```python
plt.plot(x,y)
plt.ylabel("y")
plt.xlabel("x")
plt.grid()
plt.title("Simple Line PLot")
plt.show()
```

we plot the x and y
we create 2 labels xlabel and ylabel , x and y respectively
we create grid using .grid()
then we assign a title using .title("title") 
then we print by plt.show()

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
highest_index = revenue.index(max(revenue))

explode = [0.04 if i == highest_index else 0 for i in range(len(revenue))]

plt.pie(revenue, labels = regions, autopct = "%5.2f%%", startangle = 7, explode = explode)
plt.legend()
plt.show()
```

we create 2 lists regions and revenue
we create a variable that has the index of the revenue list where the value of revenue is the greatest
in this case the index will be 0
we initialise explode where if i equals the highest revenue index (here 0) then it will be 0.04 if not then it will be zero
so explode will be 0.04,0,0,0
autopct is used to display the percentage of the pieces of the pie
