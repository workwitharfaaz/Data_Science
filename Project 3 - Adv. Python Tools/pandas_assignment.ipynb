```python
import pandas as pd
```

we import pandas

```python
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }
```

we create a dataframe "data"

```python
df = pd.DataFrame(data)
```

we convert the dictionary into pandas dataframe
using y = pd.DataFrame(x)

```python
df.head(5)
```

we print the top five rows

```python
df.describe()
```

we print summary statistics of the age and salary column using .describe()

```python
print("Average salary of HR department", df[df["Department"] == "HR"]["Salary"].mean())
```

we print the average salary of HR department by
filtering the rows where department is hr and then
from the filtered df we calculate mean of salary column using .mean.

```python

df["Bonus"] = df["Salary"] * 0.10
df.head(5)
```

we derive bonus column by multiplying salary column by 
0.10 which gives us 10 percent of the value of salary

```python
df_age_bw_25_30 = df[(df["Age"] >= 25) & (df["Age"] <= 30)]
print(df_age_bw_25_30)
```

we initialise df_age_bw_25_30 where we store all the rows where age is greater than 25 & age is less than 30

```python
grouped = df.groupby("Department")["Salary"].mean()
print("Average salary by department:")
print(grouped)
```

we initialise grouped that has mean of salary after we group df by department using .groupby

```python
asc_by_sal = df.sort_values(by="Salary")
asc_by_sal.to_csv("Sorted by Salary.csv",index=False)
```

we sort the rows in ascending order of salary using .sort_values and passing parameter by="Salary"
we save the asc_by_sal using .to_csv and parameter index=False
